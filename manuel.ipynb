```python
nombres= open("nombres_1.txt", "r", encoding="utf-8")
eval1 = open("eval1.txt", "r", encoding="utf-8")
eval2 = open("eval2.txt", "r", encoding="utf-8")
```

abro los archivos y asigno a las variables


```python
alumnos = nombres.read().split()
nota1 = eval1.read().split(",")
nota2 = eval2.read().split(",")
```

a las variables alumnos,nota1,nota2 asigo los datos de los archivos

```python
nombres.close()
eval1.close()
eval2.close()
```

cierro las variables que tenian los archivos

```python
total_suma = []
promedio = 0
```

creo las variables total_suma y promedio

```python
for nom, N1, N2 in zip(alumnos, nota1, nota2) :
    total_suma.append((nom, int(N1) + int(N2)))
    promedio += int(N1) + int(N2)
```

asigno a la lista total_suma el nombre y lu nota total del alumno y a promedio voy sumando todas las notas

```python
if promedio> 0 :
    print(f"el promedio total de las notas es:  {promedio/ len(total_suma):.2f}")
```

calculo el promedio total

```python
print("LOS ALUMNOS CON NOTA MENOR AL PROMEDIO SON: ")

for alum in total_suma :
    if alum[1] < promedio / len(total_suma) :
        print(alum[0], alum[1])
```
